```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objs as go
```

```python
df = yf.download(tickers='SPY', period='0d', interval='1m')
```

```python
df
```

```python
df['Middle Band'] = df['Close'].rolling(window=21).mean()
df['Upper Band'] = df['Middle Band'] + 1.96 * df['Close'].rolling(window=21).std()
df['Lower Band'] = df['Middle Band'] - 1.96 * df['Close'].rolling(window=21).std()
```

```python
# declare figure
fig = go.Figure()

# add traces
fig.add_trace(go.Scatter(x=df.index, y= df['Middle Band'], line=dict(color='blue', width=.7), name='Middle Band'))
fig.add_trace(go.Scatter(x=df.index, y= df['Upper Band'], line=dict(color='red', width=.7), name='Upper Band (sell)'))
fig.add_trace(go.Scatter(x=df.index, y= df['Lower Band'], line=dict(color='green', width=.7), name='Lower Band (buy)'))

# Add candlestick
fig.add_trace(go.Candlestick(
                x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'],
                name='Market Data'))

# add titles
fig.update_layout(
        title= 'SPY live share price evolution',
        yaxis_title='Stock Price (USD per shares)')

# X-Axes
fig.update_xaxes(
        rangeslider_visible = True,
        rangeselector = dict(
            buttons = list([
                dict(count=15, label='15m', step='minute', stepmode='backward'),
                dict(count=45, label='45m', step='minute', stepmode='backward'),
                dict(count=1, label='HTD', step='hour', stepmode='todate'),
                dict(count=3, label='3h', step='hour', stepmode='backward'),
                dict(step='all')
            ])
        )
)

# show plot
fig.show()
```
